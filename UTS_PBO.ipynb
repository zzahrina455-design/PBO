```python
#1. Inheritance
class Pengguna:                                                                       #Class induk
    def __init__(self, email, username, password):                                    #yang dijalankan saat objek dibuat
        self.email = email                                                            #untuk menyimpan email yang didaftarkan
        self.username = username                                                      #untuk menyimpan username yang didaftarkan
        self.password = password                                                      #untuk menyimpan password yang didaftarkan
        self.role = self.set_role()                                                   #menentukan role sesuai email

    def set_role(self):                                                               #menentukan role
        if "@admin.com" in self.email:                                                #jika email @admin.com
            return "admin"                                                            #maka role adalah admin
        else:                                                                         #jika bukan
            return "user"                                                             #maka role adalah user

    def login(self):                                                                  #fungsi login
        return (f"{self.username} login sebagai {self.role}")                         #untuk menampilkan status login

class Admin(Pengguna):                                                                #class turunan dari class induk
    def Dashboard_admin(self):                                                        #untuk menampilkan dashboard admin
        return "Masuk dashboard admin : update status antrian dan hapus data antrian" #yang bisa dilakukan admin di sistem

class Pasien(Pengguna):                                                               #class turunan dari class induk
    def Dashboard_Pasien(self):                                                       #untuk menampilkan dashboard pasien
        return "Masuk dashboard Pasien : ambil nomor antrian"                         #yang bisa dilakukan pasien di sistem

def register(email, username, password):                                              #fungsi untuk membuat akun baru
    if "@admin.com" in email:                                                         #jika email @admin.com
        return Admin(email, username, password)                                       #maka akan membuat akun admin
    else:                                                                             #jika bukan
        return Pasien(email, username, password)                                      #maka akan membuat akun pasien

Pengguna1 = register("zamz@gmail.com", "zam", "12345")                                #membuat akun user
Pengguna2 = register("Zahra@admin.com", "Zahra", "123456")                            #membuat akun admin

print(Pengguna1.login())                                                              #menampilkan status login
print(Pengguna1.Dashboard_Pasien())                                                   #akses dashboard pasien

print(Pengguna2.login())                                                              #menampilkan status login
print(Pengguna2.Dashboard_admin())                                                    #akses dashboard admin
```

```python
#2. Inheritance
class AntrianPoliklinik:                                                                   #class induk
    def __init__(self, nama_pasien, tanggal):                                              #dijalankan saat objek dibuat
        self.nama_pasien = nama_pasien                                                     #menyimpan nama pasien
        self.tanggal = tanggal                                                             #menyimpan tanggal pendaftaran
        self.nomor = None                                                                  #menyimpan nomor antrian (awal kosong)
        self.kode_poli = ""                                                                #kode poli (diisi di class turunan)

    def ambil_nomor(self, nomor):                                                          #untuk mengambil nomor antrian
        self.nomor = f"{self.kode_poli}-{nomor:03d}"                                       #format nomor (contoh: A-001)

    def info(self):                                                                        #untuk menampilkan data nomor antrian
        return f"Nama: {self.nama_pasien}\nTanggal: {self.tanggal}\nNomor: {self.nomor}"   #output data

class AntrianPoliUmum(AntrianPoliklinik):                                                  #class turunan dari class induk
    def __init__(self, nama_pasien, tanggal):                                              #yang dijalankan
        super().__init__(nama_pasien, tanggal)                                             #mengambil atribut dari class induk
        self.jenis_poli = "Poli Umum"                                                      #jenis poli
        self.jam = "08:00 - 12:00"                                                         #jam pelayanan
        self.kode_poli = "U"                                                               #kode poli untuk nomor antrian

class AntrianPoliGigi(AntrianPoliklinik):                                                  #turunan dari class induk
    def __init__(self, nama_pasien, tanggal):                                              #yang dijalankan
        super().__init__(nama_pasien, tanggal)                                             #mengambil atribut dari class induk
        self.jenis_poli = "Poli Gigi"                                                      #jenis poli
        self.jam = "09:00 - 13:00"                                                         #jam pelayanan
        self.kode_poli = "G"                                                               #kode poli

class AntrianPoliAnak(AntrianPoliklinik):                                                  #turunan dari class induk
    def __init__(self, nama_pasien, tanggal):                                              #yang dijalankan
        super().__init__(nama_pasien, tanggal)                                             #mengambil atribut dari class induk
        self.jenis_poli = "Poli Anak"                                                      #jenis poli
        self.jam = "08:00 - 11:00"                                                         #jam pelayanan
        self.kode_poli = "A"                                                               #kode poli

class AntrianPoliTHT(AntrianPoliklinik):                                                   #turunan dari class induk
    def __init__(self, nama_pasien, tanggal):                                              #yang dijalankan
        super().__init__(nama_pasien, tanggal)                                             #mengambil atribut dari class induk
        self.jenis_poli = "Poli THT"                                                       #jenis poli
        self.jam = "10:00 - 14:00"                                                         #jam pelayanan
        self.kode_poli = "T"                                                               #kode poli

class AntrianPoliPenyakitDalam(AntrianPoliklinik):                                         #turunan dari class induk
    def __init__(self, nama_pasien, tanggal):                                              #yang dijalankan
        super().__init__(nama_pasien, tanggal)                                             #mengambil atribut dari class induk
        self.jenis_poli = "Poli Penyakit Dalam"                                            #jenis poli
        self.jam = "13:00 - 16:00"                                                         #jam pelayanan
        self.kode_poli = "P"                                                               #kode poli

data1 = AntrianPoliAnak("Zam", "2026-04-13")                                               #membuat objek dari poli anak
data1.ambil_nomor(1)                                                                       #mengambil nomor antrian pertama

print(data1.info())                                                                        #menampilkan data pasien + nomor antrian
print("Poli:", data1.jenis_poli)                                                           #menampilkan jenis poli
print("Jam:", data1.jam)                                                                   #menampilkan jam pelayanan
```

```python
#3. Polimorfisme
class AntrianPoliklinik:                                            #class induk
    def ambil_nomor(self, urutan):                                  #untuk mengambil nomor antrian
        return f"Nomor antrian: {urutan}"                           #urutan nomor antrian

class AntrianPoliUmum(AntrianPoliklinik):                           #class turunan
    def ambil_nomor(self, urutan):                                  #fungsi untuk mengambil nomor
        return f"Nomor antrian: U-{urutan:03d}"                     #format khusus poli umum

class AntrianPoliGigi(AntrianPoliklinik):                           #class turunan
    def ambil_nomor(self, urutan):                                  #fungsi untuk mengambil nomor
        return f"Nomor antrian: G-{urutan:03d}"                     #format khusus poli gigi

class AntrianPoliAnak(AntrianPoliklinik):                           #class turunan
    def ambil_nomor(self, urutan):                                  #fungsi untuk mengambil nomor
        return f"Nomor antrian: A-{urutan:03d}"                     #format khusus poli anak

class AntrianPoliTHT(AntrianPoliklinik):                            #class turunan
    def ambil_nomor(self, urutan):                                  #fungsi untuk mengambil nomor
        return f"Nomor antrian: T-{urutan:03d}"                     #format khusus poli THT

class AntrianPoliPenyakitDalam(AntrianPoliklinik):                  #class turunan
    def ambil_nomor(self, urutan):                                  #fungsin untuk mengambil nomor
        return f"Nomor antrian: P-{urutan:03d}"                     #format khusus poli Penyakit Dalam

poli_list = [                                                       #list poli
    AntrianPoliUmum(),                                              #poli umum
    AntrianPoliGigi(),                                              #poli gigi
    AntrianPoliAnak(),                                              #poli anak
    AntrianPoliTHT(),                                               #poli THT
    AntrianPoliPenyakitDalam()                                      #poli Penyakit Dalam
]

for i, poli in enumerate(poli_list, start=1):                      #perulangan mulai dari nomor 1
    print(poli.ambil_nomor(i))                                     #menampilkan nomor antrian
```

```python
#4. Polimorfisme
class StatusAntrian:                                                                              #class induk
    def __init__(self, nama_pasien):                                                              #yang dijalankan
        self.nama_pasien = nama_pasien                                                            #untuk menyimpan nama pasien

    def Status(self):                                                                             #method umum
        pass                                                                                      #tidak ada isi

class AntrianAktif(StatusAntrian):                                                                #class turunan
    def Status(self):
        return f"{self.nama_pasien}: Tidak bisa daftar lagi karena status antrian masih menunggu" #menampilkan jika nomor antrian sebelumnya belum terverifikasi selesai maka tidak bisa daftar lagi


class AntrianSelesai(StatusAntrian):                                                              #class turunan
    def Status(self):
        return f"{self.nama_pasien}: Bisa daftar lagi karena status antrian sudah selesai"        #menampilkan jika nomor antrian sebelumnya sudah terverifikasi selesai maka bisa daftar lagi

data_antrian = [                                                                                  #list data status antrian
    AntrianAktif("Zam Zam"),                                                                      #antrian status aktif
    AntrianSelesai("Ros"),                                                                        #antrian status selesai
    AntrianAktif("Nafhah"),                                                                       #antrian aktif
    AntrianSelesai("Chelsea")                                                                     #antrian selesai
]

# Polimorfisme
for data in data_antrian:                                                                         #perulangan setiap objek
    print(data.Status())                                                                          #memanggil method yang sama, tapi hasil berbeda
```

```python
#5. Enkapsulasi
import hashlib                                                           #mengimport library hashlib

class Pengguna:                                                          #class induk
    def __init__(self):                                                  #yang dijalankan
        self.__password = None                                           #atribut private (tidak bisa diakses langsung dari luar class)

    def set_password(self, password):                                    #fungsi untuk menyimpan password
        self.__password = hashlib.sha256(password.encode()).hexdigest()  #password diubah menjadi hash agar lebih aman sebelum disimpan

    def cek_password(self, input_password):                              #fungsi untuk mengecek password
        input_hash = hashlib.sha256(input_password.encode()).hexdigest() #password yang dimasukkan pengguna di-hash terlebih dahulu
        return input_hash == self.__password                             #membandingkan hash input dengan hash yang tersimpan (True/False)

pengguna = Pengguna()                                                    #membuat objek dari class Pengguna
pengguna.set_password("12345")                                           #menyimpan password

print("Password benar:", pengguna.cek_password("12345"))                 #mengecek password yang benar → hasil True

print("Password salah:", pengguna.cek_password("salah"))                 #mengecek password yang salah → hasil False
```
